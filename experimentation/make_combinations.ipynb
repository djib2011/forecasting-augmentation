```python
i1 = np.random.choice(np.arange(data.shape[0]), num_samples)
i2 = np.random.choice(np.arange(data.shape[0]), num_samples)
i3 = np.random.choice(np.arange(data.shape[0]), num_samples)
i4 = np.random.choice(np.arange(data.shape[0]), num_samples)
i5 = np.random.choice(np.arange(data.shape[0]), num_samples)

combs = (data[i1] + data[i2] + data[i3] + data[i4] + data[i5]) / 5

plt.figure(figsize=(20, 20))

for e, i in enumerate(inds):
    ax = plt.subplot(4, 4, e+1)
    
    ax.plot(data[i1[i]], c='0.5', ls='--')
    ax.plot(data[i2[i]], c='0.5', ls='--')
    ax.plot(data[i3[i]], c='0.5', ls='--')
    ax.plot(data[i4[i]], c='0.5', ls='--')
    ax.plot(data[i5[i]], c='0.5', ls='--')
    ax.plot(combs[i])
```
