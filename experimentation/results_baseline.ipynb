```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
results = pd.read_csv('../reports/lstm_tuning_windows/results.csv')
```

```python
def find_exp(input_len=18, direction='bi', size=16, depth=2, ensemble=None):
    
    res = results[(results['input_len'] == input_len) & (results['direction'] == direction) & 
                  (results['size'] == size) & (results['depth'] == depth)]
    
    if ensemble is None:
        return res
    else:
        return res[res['ensemble'] == ensemble]
```

## Investigate snapshot ensembling

```python
r1 = results[results['ensemble'] == False]

sns.lineplot(x='num', y='smape', data=r1)
```

```python
r1 = find_exp(direction='bi', size=16, depth=2, ensemble=True)
```

```python
plt.plot(r1['smape'])
```
